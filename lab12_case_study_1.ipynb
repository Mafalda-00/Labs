```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('migracoes2015.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
G = nx.DiGraph() # directed
print(G)
```

```python
# G.add_edge(a,b, weight= c)

for a,b,c in zip(df.origem, df.destino, df.numeros):
    if c >= 500000:
        G.add_edge(a,b, weight= c)

G.degree(weight='weight')
```

```python
degree=G.degree()
print(degree)
```

```python
#dictionary with degrees
degree_values = dict(degree).values()
print(degree_values)
```

```python
G.nodes # see all nodes
```

```python
G.edges
```

```python
#draw chart
nx.draw(G, dim=100, with_labels=True)
```

```python
#or
plt.figure(figsize=(20, 20))
po1 = nx.spring_layout(G,scale=2)
nx.draw(G, node_size=10, pos=po1, edge_color='y', label='r', with_labels=True)
plt.show()
```

```python
#or for po1 and po1 = nx.random_layout(G)
degree =[deg for node, deg in nx.degree(G)]
max(degree)
```

```python
min(degree)
```

```python
kavg = np.mean(degree)
print(kavg)
```

```python
#convert directed to undirected
G1 = G.to_undirected()
#Coeficiente de clustering
nx.clustering(G1)
```

```python
#closeness centrality of network
nx.closeness_centrality(G)
```

```python
#eigenvector centrality of network
nx.eigenvector_centrality(G)
```

```python
#degree centrality
nx.degree_centrality(G)
```

```python
#number of linked components
nx.number_connected_components(G1)
```

```python
#node linked to component
nx.node_connected_component(G1, 'Mexico')
```
